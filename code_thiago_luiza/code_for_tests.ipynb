```python
import numpy as np

from tools import coordinates3D_MPI_v2, rank_to_coords
from tools import input_simulation_parameters
```

```python
internal_simulpar = input_simulation_parameters('simulation_input2.in')

# Vari√°veis globais
Gnx = internal_simulpar.mesh[0] 
Gny = internal_simulpar.mesh[1] 
Gnz = internal_simulpar.mesh[2]

GLx = internal_simulpar.Dom[0]
GLy = internal_simulpar.Dom[1]
GLz = internal_simulpar.Dom[2]

Px = 2
Py = 2
Pz = 1

print(f'Global domain:  [0,{GLx}]x[0,{GLy}]x[0,{GLz}]')
print(f'Global mesh:    {Gnx}x{Gny}x{Gnz}')
print(f'Subdomain grid: {Px}x{Py}x{Pz}')
```

```python
idx = []
coord = []

for rank in range(Px*Py*Pz):
    px, py, pz = rank_to_coords(rank, Px, Py, Pz)
    idx_rank, coord_rank = coordinates3D_MPI_v2(Gnx, Gny, Gnz, GLx, GLy, GLz,
                          Px, Py, Pz, rank)
    print(f'Rank {rank}')
    print(f'Process coords:    ({px},{py},{pz})')
    print(f'Local idx shape:   {idx_rank.shape}')
    print(f'Local coord shape: {coord_rank.shape}\n')
    idx.append(idx_rank)
    coord.append(coord_rank)
idx_rank, coord_rank = coordinates3D_MPI_v2(Gnx, Gny, Gnz, GLx, GLy, GLz,
                          1, 1, 1, rank)
print(f'Global idx shape:   {idx_rank.shape}')
print(f'Global coord shape: {coord_rank.shape}')
idx.append(idx_rank)
coord.append(coord_rank)
```
